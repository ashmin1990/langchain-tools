```python
from langchain_openai import ChatOpenAI
from langchain_core.tools import tool
import requests
from dotenv import load_dotenv
import os

load_dotenv()
```

```python
from langchain_community.tools import DuckDuckGoSearchRun

search_tool = DuckDuckGoSearchRun()
```

```python
api_key = os.getenv("WEATHER_STACK_API_KEY")

@tool
def get_weather_data(city: str) -> str:
  """
  This function fetches the current weather data for a given city
  """
  url = f'https://api.weatherstack.com/current?access_key={api_key}&query={city}'

  response = requests.get(url)

  return response.json()
```

```python
llm = ChatOpenAI()
```

```python
from langchain_classic.agents import create_react_agent, AgentExecutor
from langchain_classic import hub
```

```python
# Step 2: Pull the ReAct prompt from LangChain Hub
prompt = hub.pull("hwchase17/react")  # pulls the standard ReAct agent prompt
```

```python
# Step 3: Create the ReAct agent manually with the pulled prompt
agent = create_react_agent(
    llm=llm,
    tools=[search_tool, get_weather_data],
    prompt=prompt
)
```

```python
# Step 4: Wrap it with AgentExecutor
agent_executor = AgentExecutor(
    agent=agent,
    tools=[search_tool, get_weather_data],
    verbose=True
)
```

```python
# Step 5: Invoke
response = agent_executor.invoke({"input": "Find the capital of Odisha, then find it's current weather condition"})
print(response)
```
